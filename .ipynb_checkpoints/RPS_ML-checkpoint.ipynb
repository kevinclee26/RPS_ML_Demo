```python
from sklearn import tree
from sklearn.preprocessing import OneHotEncoder
import csv
import pandas as pd
```

```python
data=[]
with open('result/result.csv', 'r') as f: 
    csvfile=csv.reader(f)
    for each_row in csvfile: 
        data.append(each_row)
```

```python
df=pd.DataFrame(data, columns=['user', 'computer', 'result'])
df.head()
```

```python
X=df[['user', 'computer']]
ohencoder=OneHotEncoder()
X_transformed=ohencoder.fit_transform(X)
transformed=pd.DataFrame(X_transformed.toarray(), columns=ohencoder.get_feature_names(['user', 'computer']))
transformed['result']=df['result']
transformed
```

```python
y=df['result']
clf=tree.DecisionTreeClassifier()
clf.fit(X_transformed, y)
transformed['predicted']=clf.predict(X_transformed)
transformed
```

```python
clf.
```

```python
from IPython.display import Image
import pydotplus
dot_data=tree.export_graphviz(clf, feature_names=ohencoder.get_feature_names(['user', 'computer']))
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
